```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
from encoder import TransformerEncoder
from dataset import data
```

```python
vocab = {"[MASK]": 0}
idx = 1

for sentence, target in data:
    for word in sentence.split():
        if word not in vocab:
            vocab[word] = idx
            idx += 1
    if target not in vocab:
        vocab[target] = idx
        idx += 1

inv_vocab = {v: k for k, v in vocab.items()}
```

```python
sentence, _ = data[0]   # SAMPLE1
words = sentence.split()
```

```python
input_ids = torch.tensor([[vocab[word] for word in words]])
```

```python
model = TransformerEncoder(d_model=64, vocab_size=len(vocab))
model.eval()
```

```python
with torch.no_grad():
    _, attention = model(input_ids)

attention_matrix = attention[0].numpy()
```

```python
plt.figure(figsize=(6, 5))
plt.imshow(attention_matrix, cmap="viridis")
plt.colorbar()

plt.xticks(range(len(words)), words)
plt.yticks(range(len(words)), words)

plt.xlabel("Key Words")
plt.ylabel("Query Words")
plt.title("Self-Attention Heatmap")

plt.show()
```
